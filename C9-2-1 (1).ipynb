```python
df = pd.read_excel(r'TorontNH.xlsx')

print('Data downloaded and read into a dataframe!')
```

```python
new_header = df.iloc[0] 
df = df[1:] 
df.columns = new_header 
df.head()
```

```python
dff=df.dropna(axis=1, how='all')
dff
```

```python
# Get names of indexes for which column Borough has value Not assigned
indexNames = dff[ dff['Borough'] == 'Not assigned' ].index
# Delete these row indexes from dataFrame
dff.drop(indexNames , inplace=True)
dff.head()
```

## Droping rows with Borough unassigned

```python
dfg = dff.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()

dfg
```

## Grouping by Postcode 

```python
dfg['Neighbourhood'].replace('Not assigned',dfg['Borough'], inplace=True)
dfg
```
